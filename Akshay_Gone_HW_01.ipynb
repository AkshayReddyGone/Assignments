# Solution 1
 <b>Score provided by Linear Regression

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression

#Load the diabetes dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

#Create a Linear Regression model and fit it to the data
model = LinearRegression().fit(X, y)

#Calculate the R-squared score provided by Linear Regression
score = model.score(X, y)

print("Score provided by Linear Regression:", score)
```

# Solution 2
    
   <b>MEAN SQUARE ERROR (MSE)

```python
from sklearn.metrics import mean_squared_error

#Predict the target values
y_pred = model.predict(X)

#Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y, y_pred)

#Print the MSE
print("Mean Squared Error (MSE):", mse)
```

# Solution 3
 <b>Only BMI feature

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#Extract the 'bmi' feature
bmi_feature = X[:, 2].reshape(-1, 1)

#Create a Linear Regression model for 'bmi' feature
bmi_model = LinearRegression()
bmi_model.fit(bmi_feature, y)

#Get the slope (coefficient) and intercept of the fitting line
slope = bmi_model.coef_[0]
intercept = bmi_model.intercept_

print("Fitting line equation: y =", slope, "* bmi +", intercept)

#Plot data points and the fitting line for 'bmi' feature
plt.scatter(bmi_feature, y, label='Data points')
plt.plot(bmi_feature, bmi_model.predict(bmi_feature), color='green', label='Fitting line')
plt.xlabel('bmi')
plt.ylabel('Target')
plt.legend()
plt.title('Univariate Linear Regression with BMI feature')
plt.show()

#Calculate the score and MSE for 'bmi' feature
bmi_score = bmi_model.score(bmi_feature, y)
bmi_y_pred = bmi_model.predict(bmi_feature)
bmi_mse = mean_squared_error(y, bmi_y_pred)

print("Score for 'bmi' feature:", bmi_score)
print("MSE for 'bmi' feature:", bmi_mse)

```
